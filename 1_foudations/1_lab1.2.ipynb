## Part 1 — Setup & Create the OpenAI Client

Verify that your API key is loaded, then import the OpenAI SDK and create a client — the same setup as Lab 1.1.

```python
import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
from openai import OpenAI

openai = OpenAI()
```

## Part 2 — Step 1: Generate a Business Idea

Ask the LLM to propose a business idea that could benefit from an Agentic AI approach. Save the response — you will reference it in the next steps.

```python
messages = [{"role": "user", "content": "Could you please propose a business idea that might be worth exploring for an Agentic AI opportunity?"}]

response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

business_idea = response.choices[0].message.content
print(business_idea)
```

## Part 3 — Step 2: Identify a Pain Point

Given the business idea, ask the LLM to surface a specific pain point that an Agentic AI solution could address.

```python
messages2 = [{"role": "user", "content": f"From the business idea '{business_idea}', please propose a pain-point that might be ripe for an Agentic solution."}]

response2 = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages2
)

pain_point = response2.choices[0].message.content
print(pain_point)
```

## Part 4 — Step 3: Propose an Agentic AI Solution

Using both the business idea and the pain point from earlier steps, ask the LLM to design an Agentic AI solution.

```python
messages3 = [{"role": "user", "content": f"Could you please propose the Agentic AI solution for the pain-point '{pain_point}' for the business idea '{business_idea}'."}]

response3 = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages3
)

solution = response3.choices[0].message.content
print(solution)
```
