## Part 1 — Setup

Import the Agents SDK and load your API key from `.env`.

```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace
```

```python
load_dotenv(override=True)
```

## Part 2 — Create an Agent

Define an `Agent` with a name, system instructions, and model. The instructions shape how the agent behaves — here, it acts as a joke teller.

```python
# Make an agent with name, instructions, model

agent = Agent(name="Jokester", instructions="You are a joke teller", model="gpt-4o-mini")
```

## Part 3 — Run the Agent with Tracing

Use `Runner.run()` inside a `trace()` context to execute the agent and capture the run for debugging. Print the `final_output` when complete.

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```
