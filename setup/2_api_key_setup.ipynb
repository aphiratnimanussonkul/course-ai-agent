---
#### Create OpenAI API Key
1. Go to https://platform.openai.com
2. Register member OpenAI Platform
3. Log in
4. Go to API Keys https://platform.openai.com/api-keys
5. Click "Create new secret key"
6. Copy API Key
---
#### Create Anthropic API Key (Claude)
1. Go to https://console.anthropic.com (Anthropic Console)
2. Sign in
3. Buy usage credits
4. Go to https://platform.claude.com/settings/workspaces/default/keys
5. Create Key & Copy
---
#### Create Gemini API Key
1. Go to https://aistudio.google.com
2. Sign in via Google Account
3. Click "Get API Key" (https://aistudio.google.com/api-keys)
4. Click "Create API Key"
5. Select Project or Create new Project
6. Copy API Key
